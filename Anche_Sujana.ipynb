```python
# <Data Cleaning  and Manipulation >
## Handle Missing Values:
# Drop columns with high missing values
data = data.dropna(axis=1, thresh=len(data) * 0.7)  ## Drop columns with >30% missing values

# Identify columns with more than 30,000 null records out of 39718 total records
cols_to_drop = [col for col in data.columns if data[col].isnull().sum() > 30000]

# Drop the identified columns
data = data.drop(columns=cols_to_drop)

# Verify the changes
print(data.isnull().sum())
data.head()
```

```python
# <Data Cleaning  and Manipulation >
#data is converted to a suitable and convenient format

#formatting emp_length to numericals and float
# Define a dictionary of replacements for column emp_length
replacements = {
    r'\+ years': '',
    r'< 1 year': '0.5',
    r' years': '',
    r' year': '',
    r'n/a': '0'
}
data['emp_length'] = data['emp_length'].astype(str).replace(replacements, regex=True)
data['emp_length'] = pd.to_numeric(data['emp_length'], errors='coerce')
print('checking if emp_length column  formatted and converted string to float \n',data['emp_length'].head(10))

#formatting int_rate removing %
data['int_rate'] = data['int_rate'].astype(str).str.replace('%', '')
data['int_rate'] = pd.to_numeric(data['int_rate'], errors='coerce')
print('checking if intest rate "%" is  formatted to float \n', data['int_rate'].head(10))


```

```python
# <Data Cleaning  and Manipulation >
# Identify numeric columns
numeric_cols = data.select_dtypes(include=['int', 'float']).columns

# fill above numberical columns, Fill missing values with median for numeric columns
data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].median())

data.isnull().sum() #checking any critical null values left out
print(data.dtypes)

data.info() #  cleaned,formatted and ensured missing values imputations
```

```python
# Correlation heatmap for numerical columns
# Select only the numeric columns
numeric_cols = data.select_dtypes(include=['int','float']).columns
exclude_vars = ['collections_12_mths_ex_med','policy_code','acc_now_delinq','delinq_amnt','chargeoff_within_12_mths','tax_liens']  # Replace 'another_variable' with the actual name
filtered_numeric_cols = [col for col in numeric_cols if col not in exclude_vars]
# Create the correlation heatmap
plt.figure(figsize=(20, 24))
sns.heatmap(data[filtered_numeric_cols].corr().round(2), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap for Numeric Columns')
plt.show()
```

```python
# Calculate the correlation matrix for the numeric columns
corr_matrix = data[filtered_numeric_cols].corr()

# Create a mask to only show the upper triangle of the correlation matrix
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Create the correlation heatmap
plt.figure(figsize=(20, 24))
sns.heatmap(corr_matrix.round(2), annot=True, cmap='coolwarm', mask=mask)
plt.title('Correlation Heatmap for Numeric Columns')
plt.show()
```

```python
#<changed 'charged off' to 'defaulted" as the dataframe has been modified in column to defaulted instead charged-off# Analysis on 'charged-off' loans

charged_off = data[data['loan_status'] == 'defaulted']

# Check for driver variables like 'grade', 'dti', 'employment length', and 'annual_inc'
sns.countplot(x='grade', data=charged_off)
plt.title('Loan Grade for Charged-Off Loans')

# Distribution of debt-to-income ratio among charged-off loans
plt.figure(figsize=(10, 6))
sns.histplot(charged_off['dti'], kde=True)
```

```python
#segmentation <need to change to 'defaulted'>
# Convert loan_status to binary: 1 for 'Charged Off', 0 for others
data['loan_status_binary'] = data['loan_status'].apply(lambda x: 1 if x == 'defaulted' else 0)
state_summary = data.groupby('addr_state').agg(
    default_rate=('loan_status_binary', 'mean'),
    avg_loan_amnt=('loan_amnt', 'mean'),
    avg_dti=('dti', 'mean')
).reset_index()
plt.figure(figsize=(15, 8))
sns.barplot(x='addr_state', y='default_rate', data=state_summary)
plt.xticks(rotation=90)
plt.title('Default Rate by State')
plt.show()
```
